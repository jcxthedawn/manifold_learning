# Order of faces using ISOMAP

```python
# load data
from scipy.io import loadmat
df = loadmat('isomap.mat')
data = df['images']
```

```python
#data
```

```python
#reshape and previews of pictures
image=data.reshape(64,64,698).T
for i in range(5):
    plt.figure()
    plt.imshow(image[i],cmap = 'gray')
plt.show()
```

```python
#calculate ajdjacency matrix recording distance in l2 norm

Euc_D = np.zeros((data.shape[1],data.shape[1]))
for i in range(data.shape[1]-1):
    for j in range(i+1,data.shape[1]):
        Euc_D[i,j] = np.sqrt(np.sum((data[:,i]-data[:,j])**2))
Euc_D = Euc_D + Euc_D.T
```

```python
Euc_D.shape
# plt.imshow(Euc_D,cmap = 'gray')
# plt.show()
```

```python
#max euc distaince
Euc_D.max()
```

```python
# set an espison value and if the distance is larger than esp value, set it to zero
# construct adjacency matrix

#set a esp slightly below the max euc distance
esp = 12

def weighted_graph(Euc_D, esp):
    A = Euc_D.copy()
    A[Euc_D>=esp]=0.
    return A

# Visualize Ajancency matrix
A=weighted_graph(Euc_D, esp)
plt.imshow(A,cmap = 'gray')
plt.colorbar();
plt.show()
```

```python
#use scipy package to compute shortest path
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import shortest_path
D = shortest_path(csgraph=A, directed=False, method= 'D')
#D
```

```python
m3 = Euc_D.shape[0]
I = np.eye(m3)
u = np.ones((m3,1), dtype=int)
H = I - (1/m3) * np.dot(u,u.T)

D2=D**2
C3 = (-1/2) * H @ D2 @ H
#C3
```

```python
#eigendecomsition 
import scipy.sparse.linalg as ll

K = 2
W,V = ll.eigs(C3,k = K)
W = W.real
V = V.real
# print('selected eigenvalues:',W)
# print('selected eigenvectors:', V)
```

```python
# construct diagonal matrix for eigenvalues
wm = np.sqrt(np.diag(W))
ZT = np.dot(V, wm)
#ZT
```

```python
#plot select some faces to plot 
# if the graph does not show up or looks wierd for the first time you run, please run it again and it should fix
fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111)
ax.set_title('2D Components from Isomap of Facial Images')
x_size=2
y_size=2
for i in np.random.randint(698,size=100):
    x0 = ZT[i,0] - (x_size / 2.)
    y0 = ZT[i,1] - (y_size / 2.)
    x1 = ZT[i,0] + (x_size / 2.)
    y1 = ZT[i,1] + (y_size / 2.)
    img = image[i]
    ax.imshow(img, aspect='auto', cmap=plt.cm.gray, 
              interpolation='nearest', zorder=100, extent=(x0, x1, y0, y1));
for i in range (ZT.shape[0]):
    #print(pc[i,0], pc[i,1])
    ax.plot([ZT[i,0]], [ZT[i,1]],'o', color = 'black');
plt.show()
```

### Compare with PCA aopproach

```python
data.shape
```

```python
dim = data.shape[0] #dimension of data  4096
num = data.shape[1] #n data points 698
```

```python
#find the covariance matrix
mu = data.T-np.tile(np.mean(data.T, axis = 1),(data.T.shape[1],1)).T

Co = (1/num) * mu.T @ mu
Co.shape
```

```python
#eigendecomsition 
import scipy.sparse.linalg as ll

K = 2
W2,V2 = ll.eigs(Co,k = K)
W2 = W2.real
V2 = V2.real
# print('selected eigenvalues:',W2)
# print('selected eigenvectors:', V2)
```

```python
pc_iso = (1/np.sqrt(W2))*np.dot(mu,V2)
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111)
ax.set_title('PCA of ISOMAP Facial Images')
x_size=0.07
y_size=0.24
for i in np.random.randint(698,size=100):
    x0 = pc_iso[i,0] - (x_size / 2.)
    y0 = pc_iso[i,1] - (y_size / 2.)
    x1 = pc_iso[i,0] + (x_size / 2.)
    y1 = pc_iso[i,1] + (y_size / 2.)
    img = image[i]
    ax.imshow(img, aspect='auto', cmap=plt.cm.gray, 
              interpolation='nearest', zorder=100, extent=(x0, x1, y0, y1))
for i in range (pc_iso.shape[0]):
    #print(pc[i,0], pc[i,1])
    ax.plot([pc_iso[i,0]], [pc_iso[i,1]],'o', color = 'black');
plt.show()
```

Clearly, we can observe that Isomap preserved more distance information than PCA in high-dimension manifold
